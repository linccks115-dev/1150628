```python
# 輸入的西元年份轉換成民國年份後輸出
year = int(input("請輸入西元年份: "))
roc_year = year - 1911

# print(f"民國{roc_year}年")
# 這行負責把最後的成果漂亮地展示在螢幕上：
# f"..." (f-string 格式化字串)：
# 這是 Python 非常現代且強大的字串語法。只要在雙引號前面加上一個字母 f，引號裡面就可以放進 大括號 { }。
# {roc_year}：
# Python 看到大括號，就會自動把剛剛算出來的 roc_year 數字給「塞」進去那個位置。
# 舉例：如果前面算出 roc_year 是 115，這行就會自動組合成 "民國115年" 並透過 print() 印出來。
 
print(f"民國{roc_year}年")
print("民國" + str(roc_year) + "年")
```
