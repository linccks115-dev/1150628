```python
'''
年齡判斷一個人的生活階段：
未滿18歲未成年
未滿65歲成年
65歲以上老年
'''
'''
age = int(input("請輸入年齡: "))
if age < 18:
    print("未成年")
elif age < 65:
    print("成年") 
else:
    print("老年")
'''

'''
海象運算子（Walrus Operator）是 Python 在 3.8 版本推出的一個明星級語法符號，寫法為 :=。
之所以叫「海象運算子」，是因為如果你把眼睛向左傾斜 90 度看這個符號（: 是眼睛，= 是牙齒），它長得就像一隻有著長長獠牙的海象。
它的正式名稱叫做「指派陳述式（Assignment Expressions）」。簡單來說，它的核心功能就是：「讓你能夠在進行條件判斷或執行運算時，『順便』把值存進變數裡。」
'''

# 方案一：三元運算子巢狀嵌套（精簡成一行 🌟）
print("未成年" if (age := int(input("請輸入年齡: "))) < 18 else "成年" if age < 65 else "老年")
```

```python
# 輸入一個學生的分數，輸出該成績級別，分數級別如下：
'''
print("A" if (score := float(input("請輸入分數: "))) >= 90 else "B" if score >= 80 else "C" if score >= 70 else "D" if score >= 60 else "E"
      )
'''

'''
score = float(input("請輸入分數: "))
if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:
    print("E")
'''

'''
方案三：使用 Python 3.10+ 最新語法 match-case（最專業、最好擴充）
如果你的 Python 環境是 3.10 以上（你目前的環境是 3.14，完全支援！），
引入現代程式語言最主流的「結構化模式比對」是最好的做法：
這是目前大型專案最推崇的寫法，它拋棄了傳統 elif 臃腫的視覺感，改用 case 區塊，
不管是未來想增加 A+ 還是調整分數區間，都極度乾淨俐落。
'''
score = float(input("請輸入分數: "))
match score:
    case _ if score >= 90: print("A")
    case _ if score >= 80: print("B")
    case _ if score >= 70: print("C")
    case _ if score >= 60: print("D")
    case _: print("E")
```

```python
# 輸入五個整數，求這五個整數的和。
'''
x=int(intput("請輸入第一個整數: ")) + int(input("請輸入第二個整數: ")) + int(input("請輸入第三個整數: ")) + int(input("請輸入第四個整數: ")) + int(input("請輸入第五個整數: "))
print(f"五個整數的總和是: {x}")
''' 

total = 0

for i in range(1, 6): # i 會依序是 1, 2, 3, 4, 5
    num = int(input(f"請輸入第{i}個整數: "))
    print(f"你輸入的第{i}個整數是: {num}")
    total += num  # 這行等於 total = total + num 的意思

print(f"五個整數的總和是: {total}")
```

```python
# 求N個整數的和，N由使用者輸入。
N = int(input("請輸入整數的個數: "))
total = 0

for i in range(1, N + 1):
    num = int(input(f"請輸入第{i}個整數: "))
    total += num

print(f"這{N}個整數的總和是: {total}")
```
